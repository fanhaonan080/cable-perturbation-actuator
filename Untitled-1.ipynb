```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
data_path = r".\exo_data\20260406_1437_test0.csv"
df = pd.read_csv(data_path)

# Create time axis (relative time in seconds)
time = df['sys_time'] - df['sys_time'].iloc[0]

# Create a comprehensive figure with multiple subplots
fig, axes = plt.subplots(4, 2, figsize=(15, 12))
fig.suptitle('Actuator System Data', fontsize=16, fontweight='bold')

# 1. Actuator Angle
axes[0, 0].plot(time, df['actuator_angle'], 'b-', linewidth=1)
axes[0, 0].set_ylabel('Angle (deg)', fontweight='bold')
axes[0, 0].set_title('Actuator Angle')
axes[0, 0].grid(True, alpha=0.3)

# 2. Actuator Velocity
axes[0, 1].plot(time, df['actuator_velocity'], 'g-', linewidth=1)
axes[0, 1].set_ylabel('Velocity (deg/s)', fontweight='bold')
axes[0, 1].set_title('Actuator Velocity')
axes[0, 1].grid(True, alpha=0.3)

# 3. Actuator Torque
axes[1, 0].plot(time, df['actuator_torque'], 'r-', linewidth=1)
axes[1, 0].set_ylabel('Torque (Nm)', fontweight='bold')
axes[1, 0].set_title('Actuator Torque')
axes[1, 0].grid(True, alpha=0.3)

# 4. Motor Current
axes[1, 1].plot(time, df['mc_current'], 'm-', linewidth=1)
axes[1, 1].set_ylabel('Current (A)', fontweight='bold')
axes[1, 1].set_title('Motor Controller Current')
axes[1, 1].grid(True, alpha=0.3)

# 5. Cam Angle
axes[2, 0].plot(time, df['cam_angle'], 'c-', linewidth=1)
axes[2, 0].set_ylabel('Angle (deg)', fontweight='bold')
axes[2, 0].set_title('Cam Angle')
axes[2, 0].grid(True, alpha=0.3)

# 6. Cam Velocity
axes[2, 1].plot(time, df['cam_velocity'], 'orange', linewidth=1)
axes[2, 1].set_ylabel('Velocity (deg/s)', fontweight='bold')
axes[2, 1].set_title('Cam Velocity')
axes[2, 1].grid(True, alpha=0.3)

# 7. Disturbance Displacement
axes[3, 0].plot(time, df['disturbance_displacement'], 'darkblue', linewidth=1)
axes[3, 0].set_ylabel('Displacement', fontweight='bold')
axes[3, 0].set_xlabel('Time (s)', fontweight='bold')
axes[3, 0].set_title('Disturbance Displacement')
axes[3, 0].grid(True, alpha=0.3)

# 8. Disturbance Velocity
axes[3, 1].plot(time, df['disturbance_velocity'], 'darkgreen', linewidth=1)
axes[3, 1].set_ylabel('Velocity', fontweight='bold')
axes[3, 1].set_xlabel('Time (s)', fontweight='bold')
axes[3, 1].set_title('Disturbance Velocity')
axes[3, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
from scipy.interpolate import PchipInterpolator

# Define parameters
torque_ramp_duration = 3.0  # seconds
initial_torque = 0.5
peak_torque = 1.0

# Define torque values at key points
torque_values = [
    initial_torque,                                          
    peak_torque,                                                
    peak_torque,                                              
    initial_torque,                                                
    initial_torque
]

# Define time points (as percentage of ramp duration)
time_points = [
    0.0, 
    torque_ramp_duration * 0.01,
    torque_ramp_duration * 0.675,
    torque_ramp_duration * 0.99,
    torque_ramp_duration
]

# Create PCHIP interpolator
pchip_torque = PchipInterpolator(time_points, torque_values)

# Generate time series for visualization
time_in_torque_mode = np.linspace(0, torque_ramp_duration+1, 1000)

# Interpolate torque based on time elapsed
desired_torque = np.array([float(pchip_torque(t)) if t <= torque_ramp_duration 
                           else torque_values[-1] for t in time_in_torque_mode])

# Plot the result
plt.figure(figsize=(12, 6))
plt.plot(time_in_torque_mode, desired_torque, 'b-', linewidth=2, label='Torque Profile')
plt.plot(time_points, torque_values, 'ro', markersize=10, label='Control Points', zorder=5)

# Annotate control points
for i, (t, tau) in enumerate(zip(time_points, torque_values)):
    percentage = (t / torque_ramp_duration) * 100
    plt.annotate(f'{percentage:.1f}%', 
                xy=(t, tau), 
                xytext=(5, 10 if i % 2 == 0 else -15),
                textcoords='offset points',
                fontsize=9)

plt.xlabel('Time (s)', fontweight='bold')
plt.ylabel('Normalized Torque', fontweight='bold')
plt.title('PCHIP Torque Ramp Profile', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

print(f"Ramp duration: {torque_ramp_duration}s")
print(f"Initial torque: {initial_torque} (50%)")
print(f"Peak torque: {peak_torque} (100%)")
print(f"Control points at: {[f'{(t/torque_ramp_duration)*100:.1f}%' for t in time_points]}")
```
